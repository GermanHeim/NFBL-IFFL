```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
mpl.rcParams["axes.spines.right"] = False
mpl.rcParams["axes.spines.top"] = False
plt.style.use("seaborn-v0_8-whitegrid")
mpl.rcParams['figure.dpi'] = 100
mpl.rcParams['savefig.dpi'] = 600
```

```python
data = pd.read_csv("data_dif_curves_noise_y10.csv", header=None)
data = data.values
num_curves_total = 50000

curves_nfbl = data[:1000, :]
curves_iffl = data[int(num_curves_total/2):int(num_curves_total/2)+1000, :]
```

```python
ysum_nfbl = curves_nfbl[:, 0]
ysum_iffl = curves_iffl[:, 0]
```

```python
plt.figure()
plt.plot(ysum_nfbl, label="NFBL")
plt.plot(ysum_iffl, label="IFFL")
plt.legend(loc='upper right')
plt.ylabel('Y 10')
```

```python
plt.figure()
plt.hist(ysum_nfbl, bins=2000, alpha=0.5, label='NFBL')
plt.hist(ysum_iffl, bins=2000, alpha=0.5, label='IFFL')
plt.legend(loc='upper right')
plt.xlabel('Y 10')
plt.ylabel('Frequency')
plt.xlim([0, 10])
```

```python
curves_nfbl_2 = data[1000:2000, :]
curves_ifbl_2 = data[int(num_curves_total/2)+1000:int(num_curves_total/2)+2000, :]

ysum_nfbl_2 = curves_nfbl_2[:, 0]
ysum_iffl_2 = curves_ifbl_2[:, 0]
```

```python
plt.figure()
plt.plot(ysum_nfbl_2, label="NFBL")
plt.plot(ysum_iffl_2, label="IFFL")
plt.legend(loc='upper right')
plt.ylabel('Y 10')
```

```python
plt.figure()
plt.hist(ysum_nfbl_2, bins=2000, alpha=0.5, label='NFBL')
plt.hist(ysum_iffl_2, bins=2000, alpha=0.5, label='IFFL')
plt.legend(loc='upper right')
plt.xlabel('Y sum')
plt.ylabel('Frequency')
plt.xlim([0, 10])
```
